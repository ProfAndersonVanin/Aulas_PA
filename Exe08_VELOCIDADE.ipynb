Faça um algoritmo para prever a velocidade média de um carro, sabendo que futuramente esse programa irá  calcular também o valor da multa, se ele tiver acima da velocidade,  para isso solicite os seguintes dados: Obs. A velocidade máxima dessa estrada é 110Km/h.
- Qual o percurso percorrido: Ex. 150km
- Qual o tempo gasto nesse percurso: Ex. 1h13
- O Resultado deverá ser: 122km/h

fluxograma08.pngNeste caso o exercício irá parar por aqui no entanto o assunto das próximas aulas incluirão operadores relacionais, lógicos e condições onde iremos calcular o valor da multa.

```python
# Coletando os valores para as variáveis de entrada
distancia = float(input("Digite a distância percorrida:"))
qtde_horas = int(input("Digite a qtde de horas gastas:"))
qtde_min = int(input("Digite a qtde de minutos gastos:"))
```

```python
distancia
```

```python
# Processamento
tempo_total = qtde_horas + (qtde_min/60)
velocidade = distancia/tempo_total
```

```python
velocidade
```

```python
# Saída do programa
print("A velocidade média foi de ",round(velocidade,2)," Km/h")
```
